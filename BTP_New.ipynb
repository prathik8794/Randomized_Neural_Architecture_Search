```python
import numpy as np
import math as mh
import matplotlib.pyplot as plt
```

```python
def H(x):
  x = x+5
  return mh.exp(-10*mh.pow(x-x*x,2) + mh.pow(1-x,2))

def S(x):
  return mh.exp(2*x)

class pdf:
  def __init__(self):
    self.mu = 0
    self.sigma = 10

  def f(self,x):
    # print("--> ",(2 * mh.pow(self.sigma, 2)))
    numerator = mh.exp(-mh.pow(x - self.mu, 2) / (2 * mh.pow(self.sigma, 2)))
    denominator = mh.sqrt(2 * mh.pi) * self.sigma
    pdf_value = numerator / denominator
    if pdf_value==0:
      pdf_value = 1e-100
      # print("--> ","x: ",x,"pdf_value: ",pdf_value,"mean: ",self.mu,"sigma: ",self.sigma,numerator,denominator,-mh.pow(x - self.mu, 2) / (2 * mh.pow(self.sigma, 2)))
    return pdf_value

  def update(self,newMu,newSigma):
    self.mu = newMu
    self.sigma = max(newSigma,1e-50)
```

```python
x = np.arange(-10,10,0.01)
y = [H(i) for i in x]

pdGauss = pdf()
y_prime = [pdGauss.f(i) for i in x]

plt.plot(x,y)
plt.plot(x,y_prime)
```

```python
def I(x,gamma):
  return x
  if x>=gamma:
    return 1
  return 0
```

```python
def update_mu(X,gamma,k,pdf_function):
    func_numerator = lambda x, k, pdf_function, gamma: I((S(H(x)) ** k) / pdf_function.f(x) , gamma) * x
    func_denomerator = lambda x, k, pdf_function, gamma: I((S(H(x)) ** k) / pdf_function.f(x) , gamma)
    return calculate_expection(X,func_numerator,k,pdf_function,gamma) / (calculate_expection(X,func_denomerator,k,pdf_function,gamma))
```

```python
def update_sigma(X,gamma,k,pdf_function):
  func_numerator = lambda x, k, pdf_function, gamma: I((S(H(x)) ** k) / pdf_function.f(x) , gamma) * (x - pdf_function.mu) ** 2
  func_denominator = lambda x, k, pdf_function, gamma: I((S(H(x)) ** k) / pdf_function.f(x) , gamma)
  return calculate_expection(X,func_numerator,k,pdf_function,gamma) / (calculate_expection(X,func_denominator,k,pdf_function,gamma))
```

```python
def calculate_expection(X,func,k,pdf_function,gamma):
  arr = []
  for x in X:
    arr.append(func(x,k,pdf_function,gamma))
  sum =0
  for i in arr:
    sum +=i
  return sum/len(X)
```
